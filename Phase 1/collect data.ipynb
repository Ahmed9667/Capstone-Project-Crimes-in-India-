```python
import selenium
from selenium import webdriver
import pandas as pd
import numpy as np
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings('ignore')
import time
```

# 1 & 2 The population and literacy rate of each state

```python
driver=webdriver.Chrome()
driver.get('https://www.indiacensus.net/literacy-rate.php')
```

```python
states = []
population = []
growth = []
for i in driver.find_elements(By.XPATH,'//table[@class="pincode-tbl"]/tbody/tr/td[2]'):
    states.append(i.text)

for i in driver.find_elements(By.XPATH,'//table[@class="pincode-tbl"]/tbody/tr/td[3]'):
    population.append(i.text)
    
for i in driver.find_elements(By.XPATH,'//table[@class="pincode-tbl"]/tbody/tr/td[5]'):
    growth.append(i.text)

populations = pd.DataFrame()
populations['states']= states
populations['poopulation'] = population
populations['literacy'] = growth
populations
```

```python
populations.to_csv('E:\\Datatrained\\Castpone project\\population and litercay of states.csv' , index=False)
```

# Aread of each state

```python
driver = webdriver.Chrome()
driver.get('https://statisticstimes.com/geography/india/indian-states-area.php')
```

```python
states = []
area = []
for i in driver.find_elements(By.XPATH,'//table[@id="table_id"]/tbody/tr/td[2]'):
    states.append(i.text)

for i in driver.find_elements(By.XPATH,'//table[@id="table_id"]/tbody/tr/td[3]'):
    area.append(i.text)
    


areas = pd.DataFrame()
areas['state'] = states
areas['Area km^2'] = area
areas
```

```python
areas.to_csv('E:\\Datatrained\\Castpone project\\areas of indian states.csv' , index=False)
```
